```python
import json
import numpy as np
import pickle

import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times"
#configure plotting
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import tikzplotlib as tikz

tikz._markers._MP_MARKER2PGF_MARKER["*"] = "mystar"
```

```python
with open("dataset_list.json", "r") as f:
    dataset_list = json.load(f)

dataset_list = dataset_list[:27]

with open("10seed_result.json", "r") as f:
    collected_result = json.load(f)


METHOD_LIST = ['la', 'ep', 'vgp', 'epcvi']
PRINT_METHOD_NAME = ['LA', 'EP', 'VI', 'Ours']

METRIC_LIST = ['test_acc', 'test_loglik']

DATASET_NAME_LIST = []
for DATASET in dataset_list:
    DATASET_NAME_LIST.append(DATASET[0])


FINISHED_DATASET = []
FINISHED_DATASET_LIST = []

for i in range(len(DATASET_NAME_LIST)):
    finished = 1
    for METRIC in METRIC_LIST:
        for METHOD in METHOD_LIST:
            if len(collected_result[DATASET_NAME_LIST[i]][METHOD][METRIC]) == 0:
                finished = 0
    if finished:
        FINISHED_DATASET.append(DATASET_NAME_LIST[i])
        FINISHED_DATASET_LIST.append(i)
    
dataset_n = []

for i in FINISHED_DATASET_LIST:
    dataset_n.append(dataset_list[i][1])
    
SORTED_FINISHED_DATASET = np.array(FINISHED_DATASET)[np.argsort(dataset_n)]
SORTED_FINISHED_DATASET_LIST = np.array(FINISHED_DATASET_LIST)[np.argsort(dataset_n)]
```

```python
DATASET_LIST = list(SORTED_FINISHED_DATASET)
METHOD_LIST = ['la', 'ep', 'vgp',  'epcvi']
COLOR_LIST = ['C0', 'C1', 'C2', 'black']

METHOD_NAME_LIST = ['LA', 'EP', 'VI', 'Ours']

MARKER_LIST = ['o', 's', '^', '*']

MARKER_SIZE_LIST = [4, 3.5, 5, 6]
```

```python
def plot_single_result(metric, relative):
    
    plt.figure(figsize = (10,5))
    
    total_dataset_res = []
    
    for (fig_num, dataset) in enumerate(DATASET_LIST):
    
        cur_dataset_res = []

        for (i, method) in enumerate(METHOD_LIST):

            cur_method_total_res = []

            for j in range(5):

                cur_res = collected_result[dataset][method]["test_"+metric][j]
                cur_method_total_res.append(cur_res)

            cur_dataset_res.append(np.mean(cur_method_total_res))
        
        if relative:
            cur_dataset_res = np.array(cur_dataset_res) / max(cur_dataset_res)
        
        total_dataset_res.append(cur_dataset_res)

        for (i, method) in enumerate(METHOD_LIST):

            plt.plot(1 + fig_num + 0* (i-1.5) / 20, cur_dataset_res[i], MARKER_LIST[i], c = COLOR_LIST[i], alpha=0.5,markersize = MARKER_SIZE_LIST[i],
                    clip_on=False)

        
        plt.xlabel("Data set \#")
        plt.ylabel('Relative accuracy')

    
    total_dataset_res = np.array(total_dataset_res)
    
    mean_over_mean = np.mean(total_dataset_res, axis = 0)
    std_over_mean = np.std(total_dataset_res, axis = 0)
    

    for (i, method) in enumerate(METHOD_LIST):
        plt.axhline(y=mean_over_mean[i], color = COLOR_LIST[i], linestyle = '-')
        print(method, std_over_mean[i])
        
    
    _data = list(zip(METHOD_NAME_LIST, MARKER_LIST, MARKER_SIZE_LIST, COLOR_LIST, mean_over_mean, std_over_mean))
    for i in [3, 2, 0, 1]:
        (method_name, marker, markersize, color, mean, std) = _data[i]
        label = f"{method_name} (mean: {mean:.3}, std.dev.: {std:.3f})"
        plt.plot([1], [5], marker, color = color, markersize = markersize, label = label, linestyle = '-')
    
    if metric == 'acc':
        plt.ylim([0.86, 1.0])
    plt.legend(loc='lower right')

    
```

```python
plot_single_result('acc', True)
```
